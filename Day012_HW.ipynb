### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
Class={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[-1].text,  d.find_all('a')[-1].text)
    try:
        Class[d.find_all('em')[-1].text]['count']+=1
        Class[d.find_all('em')[-1].text]['news'].append({"title":d.find_all('a')[-1].text,"time":d.find(class_="date").text})
    except:
        Class[d.find_all('em')[-1].text]={'count':1,'news':[{"title":d.find_all('a')[-1].text,"time":d.find(class_="date").text}]}
```

### ③ 哪一個類別的文章最多

```python
Max=0
for d in Class:
    if(Class[d]['count']>Max):
        Max=Class[d]['count']
for d in Class:
    if(Class[d]['count']==Max):
        print("最多的分類為",d)
```
